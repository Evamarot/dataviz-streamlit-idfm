```python
import numpy as np
def process_gtfs(gtfs_data):
    """
    Cleans and merges GTFS data to calculate the average passage frequency 
    for each station, and extracts the line name.
    
    Outputs: DataFrame with ['station_name', 'line_name', 'stop_lat', 'stop_lon', 'avg_passages']
    """
    
    df_stops = gtfs_data['stops']
    df_stop_times = gtfs_data['stop_times']
    df_trips = gtfs_data['trips']
    df_routes = gtfs_data['routes']

    # --- 1. Link Trips to Routes to get the Line Name ---
    # Merge trips and routes on 'route_id'
    df_trips_routes = df_trips[['route_id', 'trip_id']].merge(
        df_routes[['route_id', 'route_short_name', 'route_long_name']],
        on='route_id',
        how='left'
    )
    
    # Prioritize the short name for the line name
    df_trips_routes['line_name_clean'] = df_trips_routes['route_short_name'].fillna(df_trips_routes['route_long_name']).astype(str)
    
    # Filter to keep only the necessary columns
    df_trips_routes = df_trips_routes[['trip_id', 'line_name_clean']]


    # --- 2. Calculate Passages Count per Stop and Hour ---
    
    # 2a. Time conversion function (handling 24+ hour format)
    def time_to_hour(time_str):
        try:
            h, m, s = map(int, time_str.split(':'))
            return h * 3600 + m * 60 + s # Total seconds from midnight
        except:
            return np.nan
            
    df_stop_times['arrival_time_sec'] = df_stop_times['arrival_time'].apply(time_to_hour)
    df_stop_times.dropna(subset=['arrival_time_sec'], inplace=True)
    
    # 2b. Extract Hour (0-23)
    # Use modulo 24 to get a standard hour of day (0-23)
    df_stop_times['hour'] = (df_stop_times['arrival_time_sec'] // 3600).astype(int) % 24 

    # 2c. Link stop_times to trip_routes to know the line for each passage
    df_frequency_raw = df_stop_times.merge(df_trips_routes, on='trip_id', how='left')
    df_frequency_raw.dropna(subset=['line_name_clean'], inplace=True) # Drop passages we can't link to a line
    
    # 2d. Count passages per stop, per hour, per line
    df_frequency = df_frequency_raw.groupby(['stop_id', 'line_name_clean', 'hour']).size().reset_index(name='passages_count')


    # --- 3. Final Aggregation and Normalization ---
    
    # 3a. Merge with 'stops' for station names and coordinates
    df_stops_clean = df_stops[['stop_id', 'stop_name', 'stop_lat', 'stop_lon']].copy()
    
    # 3b. Normalize station names (CRUCIAL FOR AIR QUALITY JOIN)
    df_stops_clean.rename(columns={'stop_name': 'station_name'}, inplace=True)
    df_stops_clean['station_name'] = df_stops_clean['station_name'].str.lower().str.strip() 

    # 3c. Calculate Average Frequency per Station across all hours/lines
    # We calculate the average frequency of all lines going through a single stop_id
    df_final_gtfs = df_frequency.merge(df_stops_clean, on='stop_id', how='left')
    
    df_station_avg_freq = df_final_gtfs.groupby(['station_name']).agg(
        stop_lat=('stop_lat', 'first'),
        stop_lon=('stop_lon', 'first'),
        avg_passages=('passages_count', 'mean'), # Average frequency score
        line_name=('line_name_clean', lambda x: ', '.join(x.unique())) # List all lines serving the station
    ).reset_index()

    return df_station_avg_freq
```

```python
# Assuming gtfs_data_raw was loaded in a previous cell

print("--- Processing GTFS data... This may take a moment due to the large size (10M+ rows) ---")
df_gtfs_processed = process_gtfs(gtfs_data_raw)

print("--- Processed GTFS Data Check ---")
print(f"Shape: {df_gtfs_processed.shape}")
print(f"Columns: {df_gtfs_processed.columns.tolist()}")

# Check the cleaned station name, coordinates, and the calculated frequency
print("\nSample (Check Clean Name, Geo, and Avg Frequency):")
print(df_gtfs_processed.head())

# Check extreme values
print("\nStations with the highest average traffic frequency:")
print(df_gtfs_processed.sort_values(by='avg_passages', ascending=False).head())
```

Join the two datasets

```python
import unicodedata
import re

def normalize_station_name(name):
    """Removes accents, converts to lowercase, and removes punctuation/spaces."""
    if pd.isna(name):
        return name
    
    name = name.lower().strip()
    name = unicodedata.normalize('NFD', name)
    name = name.encode('ascii', 'ignore').decode('utf-8')
    name = re.sub(r'[^\w]', '', name).strip()
    return name
```

```python
df_air_processed['station_name_clean'] = df_air_processed['station_name'].apply(normalize_station_name)
df_air_processed
```

```python
df_gtfs_processed['station_name_clean'] = df_gtfs_processed['station_name'].apply(normalize_station_name)
df_gtfs_processed
```

```python
df_merged = pd.merge(
    df_air_processed,
    df_gtfs_processed,
    on='station_name_clean', 
    how='inner',
    # Add suffixes just in case, but 'stop_lat' and 'stop_lon' are only in GTFS side now
    suffixes=('_air', '_gtfs') 
)
```

```python
print(f"Merge successful: {df_merged.shape[0]} stations were successfully matched.")
```
